### this script preprocess the Gaussian input com file parameter to the desired value (regular species version)
### this should be the first script to run

### put all the com files and sdf files into origin_tom subfolder under the same directory as this script

```python
origin_folder = "origin_tom"
new_folder = "closed_shell_tom"
```

```python
# grep all com files in the current directory
import os, re

# list all files in the origin_tom subdirectory
files = [f for f in os.listdir(origin_folder) if os.path.isfile(os.path.join(origin_folder, f)) and f.endswith(".com")]
# sort the list by this regex int(re.search(r"^\D+(\d+)_{1}", x).group(1))
files.sort(key=lambda x: int(re.search(r"^\D+(\d+)_{1}", x).group(1)))
files
```

```python
# for each file, use regex to search for this string "pop=nbo7" and swap it to pop=NBO7read
# then read line by line and look for string "--Link1--"
# after the first time we see this string, continue to read line by line and until we see string "0 1"
# first check if the next next line is "$nbo nrt $end"
# if not add a new line after this line, and add another newline with "$nbo nrt $end"
# make a new folder name regular_preprocessed and save the new files there
os.makedirs(new_folder, exist_ok=True)

for file in files:
    first_match = True
    check_flag = False

    with open(os.path.join(origin_folder, file), "r") as f:
        lines = f.readlines()
        print(f"---------------------------------------------")
        print(f"reading file {file}")
        for index, line in enumerate(lines):
            if "pop=nbo7" in line:
                lines[index] = line.replace("pop=nbo7", "pop=NBO7read")
                print(f"found pop=nbo7 at index {index}, changing it to pop=NBO7read")
            if "--Link1--" in line:
                if first_match:
                    first_match = False
                    check_flag = True
                    print(f"found --Link1-- first time at index {index}")
            if check_flag:
                if "0 1" in line:
                    print(
                        f"found 0 1 at index {index} while checking flag is {check_flag}"
                    )
                    if "$nbo nrt $end" not in lines[index + 2]:
                        # add a new line after this line
                        lines.insert(index + 1, "\n")
                        lines.insert(index + 2, "$nbo nrt $end\n")
                        print(f"adding $nbo nrt $end at index {index + 2}")
                    check_flag = False
                    break
                
    # save the new file into regular_preprocessed folder with the same name
    with open(os.path.join(new_folder, file), "w") as f:
        f.writelines(lines)
        print(f"writing new file {file} into regular_preprocessed folder")
```
