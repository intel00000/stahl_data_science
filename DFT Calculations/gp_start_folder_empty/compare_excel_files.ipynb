```python
import pandas as pd

def compare_excel_files(file1, file2):
    # Load both Excel files
    excel1 = pd.ExcelFile(file1)
    excel2 = pd.ExcelFile(file2)
    
    # Check if both files have the same sheets
    if set(excel1.sheet_names) != set(excel2.sheet_names):
        print("The Excel files have different sheets.")
        return False
    
    match = True
    # Loop through each sheet and compare
    for sheet in excel1.sheet_names:
        df1 = pd.read_excel(excel1, sheet_name=sheet)
        df2 = pd.read_excel(excel2, sheet_name=sheet)
        
        # Check if the dataframes are the same
        if not df1.equals(df2):
            print(f"Mismatch found in sheet: {sheet}")
            match = False
            # Optionally, show the differences
            diff = df1.compare(df2)
            print(f"Differences in {sheet}:\n{diff}")
    
    if match:
        print("All cells match across all worksheets.")
    else:
        print("Some cells do not match.")

    return match

# Usage
file1 = 'pyrd_properties_postprocessed_for_pyrd1_to_pyrd13_old.xlsx'
file2 = 'pyrd_properties_postprocessed_for_pyrd1_to_pyrd13.xlsx'

compare_excel_files(file1, file2)
```
